```python
import requests
import pandas as pd
from io import StringIO

# File ID
file_id = "12S9bS5TKQlB3RMrYW32HgJ-FOWNenZA1"

# Direct download endpoint (bypasses warning)
download_url = f"https://drive.usercontent.google.com/download?id={file_id}&export=download"

# Stream content
response = requests.get(download_url, stream=True)
response.raise_for_status()  # check for errors

# Convert to pandas
file_content = StringIO(response.content.decode('utf-8'))
df = pd.read_csv(file_content)

print(df.head())
```

```python


from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
import pandas as pd
import io

# 🔑 Authenticate
auth.authenticate_user()
gauth = GoogleAuth()
gauth.LocalWebserverAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)

# 📂 File ID from your share link
file_id = "12S9bS5TKQlB3RMrYW32HgJ-FOWNenZA1"

# ⚡ Get file content as a string
downloaded = drive.CreateFile({'id': file_id})
content = downloaded.GetContentString()   # << no saving, direct content

# 🐼 Load directly into pandas
df = pd.read_csv(io.StringIO(content))

df.head()
```
